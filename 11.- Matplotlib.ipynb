```python
#Matplotlib es una libreria para graficar en 2 y e D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#grafico de lineas
#generar 100 valores entre 0 y 5 unado numpy
x=np.linspace(0,5,100)#linspace crea una serie numeria entre los dos valores con la cantidad de elementos del tercer dato
print(x)
```

```python
figure,ax=plt.subplots(2,2)#se pueden recoger mas de un dato por funcion
print(ax)#ax tendra los datos de los cuadrantes
print(figure)#figuere es el resultado de plano cartesiano, 2 positivos y 2 negativos
```

```python
print(ax[0,0])#muestra la data de primer cuadrante
```

```python
fig, ax = plt.subplots(1, 1)#cuadrante positivo
ax.plot(x, x**2)#ingresar los datos aal grafico, notar que para cada valor de x, primer parametro x cuadrado ser√° y
plt.show()#mostrar
```

```python
#%matplotlib inline evita llamar a plt.show siempre
#labels
ax.set_xlabel('X')
ax.set_ylabel('c cuadrado')
display(fig)
```

```python
# m√°s de un gr√°fico por dibujo
fig, ax = plt.subplots(1, 1)

ax.plot(x, x, label='$x$')#asignar series #curva x=y
ax.plot(x, x * x, label='$x^2$')#y=x cuadrado
ax.plot(x, x**3, label='$x^3$')#y=x cubo

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Polynomials')

ax.grid(True)#poner grid al grafico

ax.legend(loc='lower left')
```
