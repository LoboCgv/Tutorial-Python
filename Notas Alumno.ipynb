```python
import numpy as np
```

```python
data=np.random.random([30,5])*100
```

```python
print(data)
```

```python
#convertir datos a nota
for i in range(30):
    #recorrer cada columna para redonder el valor
    for j in range(5):
        data[i,j]=round(data[i,j])#redondea al entero superior cuando es mayor o igual a .5 el decimal
```

```python
def getNota(valor):
    if(valor<60):
        nota=(3)*(valor/60)+1
    else:
        nota=(3)*((valor-60)/40)+4
    return nota
#obtener notas segubn algoritmo de distribucion
res=np.copy(data)
for i in range(30):
    #recorrer cada columna para redonder el valor
    for j in range(5):
        res[i,j]=getNota(data[i,j])
print(res)

```

```python
#promedio por certamen
for i in range(5):
    print("certamen ",i," promedio: ",np.mean(res[:,i]))
#promedio por alumno
promedios=np.zeros(30)
for i in range(30):
    print("alumno ",i, "promedio",np.mean(res[i,:]))
    promedios[i]=np.mean(res[i,:])
```

```python
aprobados=promedios>=4
print("aprobados",promedios[aprobados])
reprobados=promedios<4
print("reprobados",promedios[reprobados])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 1)#cuadrante positivo

plt.scatter(range(len(promedios[aprobados])),promedios[aprobados],c="blue")
plt.scatter(range(len(promedios[reprobados])),promedios[reprobados],c="red")
plt.scatter(range(len(promedios)),promedios,c="green")
plt.show()#mostrar
```
