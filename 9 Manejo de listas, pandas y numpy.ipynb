```python
import numpy as np
```

```python
#creacvion de listas
velocidad=[2,15,6,7,17,7]
distancia=[3,20,6,10,11,2]
zipeado=list(zip(velocidad,distancia))
npVelocidad=np.array(velocidad)
npDistancia=np.array(distancia)
print(zipeado[0][0])
```

```python
npVelocidad.shape
print(npVelocidad.sum())
def promedio(npArray):#para una dimension
    return npArray.mean()#obtiene promedio
print(promedio(npVelocidad))
print("velocidad bajo promedio ",list(filter(lambda x:x<promedio(npVelocidad),velocidad)))
#x=zipeado[0]
#print(x[0])
print("velocidad bajo promedio y distancia bajo promedio",list(filter(lambda x: x[0]<promedio(npVelocidad) and x[1]<promedio(npDistancia),zipeado)))
```

```python
auto1 = ['Mazda RX4', 21.0, 6, False, 4]
auto2 = ['Merc 240D', 24.4, 4, True, 2]
auto3 = ['Merc 280', 9.2, 6, True, 4]
auto4= ['Valiant', 18.1, 6, True, 1]
auto5 = ['Merc 450SL', 17.3, 8, False, 3]
auto6 = ['Toyota Corolla', 33.9, 4, True, 1]
```

```python
autos=[auto1,auto2,auto3,auto4,auto5,auto6]
def isNumero(x):
    if isinstance(x,int) or isinstance(x,float) or isinstance(x,complex):
        if(x==True or x==False):
            return False
        else:
            return True
    else:
        return False
```

```python
def soloNumeros(lista):
    result=[]
    for i in range(len(lista)):
        result+=list(filter(isNumero,lista[i]))
    return result
        #print(list(map(isNumero,autos[:])))
```

```python
npArray=np.array(soloNumeros(autos))
print(npArray)
print(promedio(npArray))
```

```python
average=promedio(npArray)
#float mayores que la media
flotantes=list(filter(lambda x:isinstance(x,float),soloNumeros(autos)))
print(list(filter(lambda x:x>average,flotantes)))
```

```python
regiones_file='data/chile_regiones.csv'
provincias_file='data/chile_provincias.csv'
comunas_file='data/chile_comunas.csv'
#leer cada uno de los archivos 
regiones=pd.read_csv(regiones_file, header=0, sep=',')#header sera la primera fila y el separador es la ,
provincias=pd.read_csv(provincias_file, header=0, sep=',')
comunas=pd.read_csv(comunas_file, header=0, sep=',')
```

```python
#viendo los valores
print('regiones: ', regiones.columns.values.tolist())
print('provincias: ', provincias.columns.values.tolist())
print('comunas: ', comunas.columns.values.tolist())
print("##################")
print(regiones.head())
print(provincias.describe())#muestra estadisticas
print(comunas.describe())
```

```python
#armando la bd
regiones_provincias=pd.merge(regiones, provincias, how='outer')
regiones_provincias.head()
provincias_comunas=pd.merge(provincias, comunas, how='outer')
provincias_comunas.head()
regiones_provincias_comunas=pd.merge(regiones_provincias, comunas, how='outer')
regiones_provincias_comunas.index.name='ID'
regiones_provincias_comunas.head()
```

```python
regiones_provincias_comunas.to_csv('data/Xchile_demographic_merge.csv', index=False)
```
