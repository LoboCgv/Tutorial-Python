```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
def getNota(valor):
    if(valor<60):
        nota=(3)*(valor/60)+1
    else:
        nota=(3)*((valor-60)/40)+4
    if nota<0:
        nota=1
    return nota

unidades=3
horasS=720
data=np.zeros(unidades+1)
tiempos=[0,horasS/unidades,2*horasS/unidades,horasS]
escala=0.6
minimo=np.copy(tiempos)*escala
app=np.random.random(unidades+1)

#app=[0.7,0.3,0.1,1]
for i in range(len(tiempos)):
    data[i]=(tiempos[i])*((i+1)**math.log(app[i],2))
    print(data[i]," ",app[i])
fig, ax = plt.subplots(1, 1)
ax.plot(data,c='r')
ax.plot(tiempos,c='b')
ax.plot(minimo,c='y')
plt.show()
fig.savefig("ensayo")
```
