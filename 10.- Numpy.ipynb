```python
#numpy es una libreria que permitirá trabajar con arreglos de una o más dimensiones de manera simple
import numpy as np# se importa la libreria completa de numpy baso el alias np
# en caso de no funcionar es necesario instalar la libreria
#pip install numpy
#conda install numpy
```

```python
#crear un vector de 10 ceros
vector=np.zeros(10)#notese que los ceros son creados en formato float
print('vector',vector)
#crear una matriz de 2X4
matriz=np.zeros([2,4])
print(matriz)
#crear una matriz de 5X5 con valores aleatorios
print("********************")
matriz=np.random.random([5,5])#notar que los valores estan entre 0 y 1
print(matriz)
#crear una matriz de 5X5 con valores entre 1 y 100
print("********************")
matriz=np.random.random([5,5])*100
print(matriz)
```

```python
#obtener datos del array
print(type(matriz))
print(matriz.shape)#obtiene las dimensiones
```

```python
#suma de matrices. Es necesario que las dimensiones sean iguales
mt1=np.random.random([3,3])*10
#print(mt1)
mt2=np.random.random([3,3])*10
res=mt1+mt2
print(res)
print("**********************")
#la multiplicacion se realiza con escalares y matrices, en el cso de las matrices
#la cantidad de filas de la primera matriz debe ser igual a la cantidad de columnas de la segunda
mt1=np.array([[1,2,3,4],[6,7,8,3]])#matriz de 2X 4
mt2=np.array([[1,3],[5,7],[6,8],[8,3]])#matriz 4X2
print(mt1*2)
print("***********")
#print(mt2*mt1)
```

```python
#acceso a filas o columnas
print(mt1[:,0])#muestra la primera columna
print(mt1[0,:])#muestra la primera fila
```

```python
#con numpy es posible trabajar usando mascaras que permitan filtrar contenidos

matriz=np.random.random([5,6])*100
masc=matriz>70
print(matriz)
print("*******************")
print(matriz[masc])#sólo imprime los vares que cumplen con la condición
```

```python
#sumar los valores de una fila
print(np.sum(matriz,axis=1))
#sumar valores de las columnas
print(np.sum(matriz,axis=0))
```

```python
#matriz transpuesta
transpuesta=matriz.T
print(transpuesta)
```

```python
#obtener promedio y desviación estandar de un conjunto de datos
data=(np.random.random(1000))#generar un vector de mil datos
print(np.mean(data))
print (np.std(data))
print(data)
```
