```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print(np.random.random)
import math
fig, ax = plt.subplots(1, 1)
data=np.zeros(720)

nota=np.random.random(720)
for i in range(720):
    data[i]=i*((i+1)**math.log(nota[i],2))
ax.plot( range(720),data,c='y')
nota=np.random.random(720)
for i in range(720):
    data[i]=i*((i+1)**math.log(0.8,2))
ax.plot( range(720),data,c='b')

plt.show()
```

```python
#considerando 10 unidades o competencias
tam=10
fig, ax = plt.subplots(1, 1)
data=np.zeros(tam)
for i in range(tam-1):
    data[i+1]=140*((i)**math.log(1,2))    
ax.plot(data,c='y')

plt.show()
```

```python
#unidades
unidades=3
horasS=720
data=np.zeros(unidades+1)
tiempos=[0,horasS/unidades,2*horasS/unidades,horasS]
escala=0.6
minimo=np.copy(tiempos)*escala
app=np.random.random(unidades+1)
#app=[0.6,0.7,0.3,0.8,1]
for i in range(len(tiempos)):
    data[i]=(tiempos[i])*((i+1)**math.log(app[i],2))
    print(data[i]," ",app[i])
fig, ax = plt.subplots(1, 1)
ax.plot(data,c='r')
ax.plot(tiempos,c='b')
ax.plot(minimo,c='y')
plt.show()
```
